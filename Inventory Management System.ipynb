**Store products and quantities**

```python
# function for storing products and quantities
products = {}

def add_product():
  try:
    product_name = str(input("Enter product name   "))
     # Check that the name contains only letters
    if not product_name.replace(" ", "").isalpha():
        raise ValueError("Name must contain only letters.")

    quantity = int(input("Enter available quantity(s)   "))
    # to check if quantity is an integer
    if quantity < 0:
        print("Invalid input   ")
    else:
        print("product record created successfully")

    products[product_name] = {
        "quantity": quantity
        }

  except ValueError as e:
        print("Error:", e)
```

**Update Product Quantity/Quantities**

```python
# function to update product quantities
def update_product():
  try:
    product_name = str(input("Enter Product's name   "))
    # check that the name contain only letters
    if not product_name.replace(" ", "").isalpha():
        raise ValueError("Name must contain only letters.")

    # to check if the product record exist
    if product_name not in products:
        raise KeyError("Record not found")

    new_quantity = int(input("Enter the Product's new quantity   "))
    # to check if the new quantity is an integer
    if new_quantity < 0:
      print("Invalid input   ")
    else:
      print(f"{product_name}'s quantity has been successfully updated")

    products[product_name] = {
        "quantity": new_quantity
        }

  except ValueError as e:
      print("Error:", e)
  except KeyError as e:
      print("Error:", e)
```

```python
update_product() # to check the function
```

```python
products # to check the function
```

**Search Products**

```python
# function to search for products
def search_product():
  try:
    product_name = str(input("Enter Product's name    "))
    # check that the name contain only letters
    if not product_name.replace(" ", "").isalpha():
        raise ValueError("Name must contain only letters.")

    # to check if the product record exist
    if product_name not in products:
        raise KeyError("Product not found")

    if product_name in products:
      print(f"{product_name} found with quantity: {products[product_name]}")

  except ValueError as e:
      print("Error:", e)
  except KeyError as e:
      print("Error:", e)
```

```python
search_product() # to check the function
```

**Remove Products**

```python
# remove product from the products record
def remove_product():
  try:
    product_name = str(input("Enter Product's name    "))

    # check that the name contain only letters
    if not product_name.replace(" ", "").isalpha():
        raise ValueError("Name must contain only letters.")

    # to check if the product record exist
    if product_name not in products:
        raise KeyError("Product not found")

    if product_name in products:
      del products[product_name]
      print("product successfully deleted")

  except ValueError as e:
      print("Error:", e)
  except KeyError as e:
      print("Error:", e)
```

```python
remove_product() # to call the function
```

```python
products # to check the function
```

**Inventory Statistics**

```python
# Inventory statistics function
quantity = []
def inventory_statistics():
  for qty in products.values():
    quantity.append(qty["quantity"])

  Total = sum(quantity)
  Average = Total/len(quantity)
  Highest = max(quantity)
  Lowest = min(quantity)

  print("Inventory Statistics")
  print("Total quantity:", Total)
  print("Average quantity:", Average)
  print("Highest quantity:", Highest)
  print("Lowest quantity:", Lowest)
```

**Indentifying Low-Stock Item**

```python

# low stock item function
def lowest_item():
  lowest_product = min(products, key=lambda product: products[product]["quantity"])
  print(
    f"Lowest stock product: {lowest_product}, "
    f"quantity: {products[lowest_product]['quantity']}"
)
```
